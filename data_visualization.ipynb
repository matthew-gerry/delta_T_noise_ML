A notebook for exploring and visualizing the Delta T shot noise data we have

```python
import numpy as np
# import matplotlib
# matplotlib.rcParams.update({'font.size': 11})
import matplotlib.pyplot as plt
import pandas as pd

# Define constants
g0 = 7.748092*10**(-5) # C^2/J/s, conductance quantum
kB = 1.380649*10**(-23) # J/K, Boltzmann constant
```

```python
# Read delta T shot noise data
df = pd.read_csv('../GNoiseData_complete.csv')

# Add rescaled data to dataframe (G already scaled by G0)
df['DeltaT_scaled'] = df['DeltaT']/df['T']
df['S_scaled'] = df['S']/(g0 * kB * df['T'])
```

Different approaches towards visualizing the data

```python
# All the data, colour scale set by deltaT/T

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["DeltaT_scaled"])
plt.xlim([0,8.3])
plt.ylim([0, 2.5])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
# plt.legend()
plt.colorbar(label="$\Delta T/T$")
plt.show()

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["DeltaT_scaled"])
plt.xlim([0,1])
plt.ylim([0, 0.25])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
# plt.legend()
plt.colorbar(label="$\Delta T/T$")
plt.show()
```

```python
# Scatter of G and S_scaled with colour scale set by unscaled, rather than scaled, Delta T values

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["DeltaT"])
plt.xlim([0,8.3])
plt.ylim([0, 2.5])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
plt.colorbar(label="$\Delta T$")
plt.show()

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["DeltaT"])
plt.xlim([0,1])
plt.ylim([0, 0.25])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
plt.colorbar(label="$\Delta T$")
plt.show()
```

```python
# Scatter plots of G and S_scaled with the average temperature T setting the colour scale rather than delta T

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["T"])
plt.xlim([0,8.3])
plt.ylim([0, 2.5])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
plt.colorbar(label="$T$")
plt.show()

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["T"])
plt.xlim([0,1])
plt.ylim([0, 0.25])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
plt.colorbar(label="$T$")
plt.show()
```
