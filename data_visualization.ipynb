A notebook for exploring and visualizing the Delta T shot noise data we have

```python
import numpy as np
# import matplotlib
# matplotlib.rcParams.update({'font.size': 11})
import matplotlib.pyplot as plt
import pandas as pd

# Set pyplot global params to get nice fonts
plt.rcParams["mathtext.fontset"] = "cm" # Match Latex font as closely as possible
plt.rcParams['font.size'] = '14' # Globally set font size of plots
plt.rcParams["font.family"] = "serif"
plt.rcParams['font.serif'] = ['cmr10']
plt.rcParams['axes.unicode_minus'] = False # NEEDED FOR MINUS SIGNS TO RENDER PROPERLY but can't give me back that hour of my life in 2021

# Define constants
g0 = 7.748092*10**(-5) # C^2/J/s, conductance quantum
kB = 1.380649*10**(-23) # J/K, Boltzmann constant
```

```python
# Read delta T shot noise data
df = pd.read_csv('../GNoiseData_complete.csv')

# Add rescaled data to dataframe (G already scaled by G0)
df['DeltaT_scaled'] = df['DeltaT']/df['T']
df['S_scaled'] = df['S']/(g0 * kB * df['T'])
```

Different approaches towards visualizing the data

```python
# All the data, colour scale set by deltaT/T

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["DeltaT_scaled"])
plt.xlim([0,8.3])
plt.ylim([0, 2.5])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
# plt.legend()
plt.colorbar(label="$\Delta T/T$")
plt.show()

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["DeltaT_scaled"])
plt.xlim([0,1])
plt.ylim([0, 0.25])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
# plt.legend()
plt.colorbar(label="$\Delta T/T$")
plt.show()
```

```python
# Scatter of G and S_scaled with colour scale set by unscaled, rather than scaled, Delta T values

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["DeltaT"])
plt.xlim([0,8.3])
plt.ylim([0, 2.5])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
plt.colorbar(label="$\Delta T$")
plt.show()

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["DeltaT"])
plt.xlim([0,1])
plt.ylim([0, 0.25])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
plt.colorbar(label="$\Delta T$")
plt.show()
```

```python
# Scatter plots of G and S_scaled with the average temperature T setting the colour scale rather than delta T

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["T"])
plt.xlim([0,8.3])
plt.ylim([0, 2.5])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
plt.colorbar(label="$T$")
plt.show()

plt.scatter(df["G"], df["S_scaled"], s=0.4, c=df["T"])
plt.xlim([0,1])
plt.ylim([0, 0.25])
plt.xlabel("$G/G_0$")
plt.ylabel("$S/G_0k_BT$")
plt.colorbar(label="$T$")
plt.show()
```

```python
# Scatter plots of G and S_scaled with the average temperature T setting the colour scale
# Showing the low-G region as a zoomed inset

# Remove points with too small delta T from the dataset
df = df[df["DeltaT"]>=1]

fig, ax = plt.subplots()

sctr = ax.scatter(df["G"], df["S_scaled"], s=0.4, c=df["T"])
ax.set_xlim([0,8.3])
ax.set_ylim([0, 1.9])
ax.set_xlabel("$G/G_0$", fontsize=24)
ax.set_ylabel(r"$S/G_0k_B\bar{T}$", fontsize=24)
ax.set_xticks([])
ax.set_yticks([])
cbar = plt.colorbar(sctr)
cbar.set_label(label=r"$\bar{T}\; (K)$", size=24)
cbar.set_ticks([])
# plt.show()

x1, x2, y1, y2 = 0.0, 1.0, 0.0, 0.25 # Boundaries of inset

axins = ax.inset_axes([0.07, 0.45, 0.45, 0.5], xlim=(x1, x2), ylim=(y1, y2), xticklabels=[], yticklabels=[])
# axins = ax.inset_axes([0.05, 0.5, 0.4, 0.45], xlim=(x1, x2), ylim=(y1, y2), xticklabels=[0.2, 0.4, 0.6, 0.8], yticklabels=[0.05,0.1,0.15,0.2])
axins.scatter(df["G"], df["S_scaled"], s=0.4, c=df["T"])
# plt.xlim([0,1])
# plt.ylim([0, 0.25])
# plt.xlabel("$G/G_0$")
# plt.ylabel("$S/G_0k_BT$")
# plt.colorbar(label="$T$")
ax.indicate_inset_zoom(axins, edgecolor='k')
plt.show()
```

Below we plot some histograms of the predictions made by our model, when trained on synthetic data, on both the synthetic data and the real data.

```python
# Import training (synthetic) and testing (real) data with predicted delta T values, identify the sets of independent Delta T values

df_synth = pd.read_csv('../training_data_with_prediction_sample.csv')
df_real = pd.read_csv('../testing_data_with_prediction_sample.csv')

unique_dT_vals_s = np.sort(np.unique(df_synth['DeltaT']))
unique_dT_vals_r = np.sort(np.unique(df_real['DeltaT']))
```

```python
unique_dT_vals_r
```

```python
# Plot a histogram of predicted Delta T values given a true value - synthetic (training) data

# Choose a set of indices at which to select out unique delta T values
dT_indices = [0, 6, 12, 18, 24, 29]

fig = plt.figure(figsize=(13, 7))
axs = fig.subplots(2,3)
for i in range(len(dT_indices)):
    dT = unique_dT_vals_s[dT_indices[i]]

    prediction_mean = df_synth.loc[df_synth['DeltaT']==dT, 'DeltaT_predicted'].mean()

    hg = axs[int(i>2),i%3].hist(df_synth.loc[df_synth['DeltaT']==dT, 'DeltaT_predicted'], bins=50, density=True)
    axs[int(i>2),i%3].vlines([dT, prediction_mean], 0, 1.1*max(hg[0]), color=['red', 'green'])
    axs[int(i>2),i%3].set_ylim([0, 1.1*max(hg[0])])
    axs[int(i>2),i%3].set_title("$\Delta T = "+str(round(dT, 1))+r", \langle \Delta T_{pred}\rangle = $"+str(round(prediction_mean, 1)))

fig.suptitle("Synthetic data - performance on training set")
plt.show()
```

```python
# Now the same for the test data

# Choose a set of indices at which to select out unique delta T values
dT_indices = [1, 2, 4, 6, 8, 10]

fig = plt.figure(figsize=(13, 7))
axs = fig.subplots(2,3)
for i in range(len(dT_indices)):
    dT = unique_dT_vals_r[dT_indices[i]]

    prediction_mean = df_real.loc[df_real['DeltaT']==dT, 'DeltaT_predicted'].mean()

    hg = axs[int(i>2),i%3].hist(df_real.loc[df_real['DeltaT']==dT, 'DeltaT_predicted'], bins=50, density=True)
    axs[int(i>2),i%3].vlines([dT, prediction_mean], 0, 1.1*max(hg[0]), color=['red', 'green'])
    axs[int(i>2),i%3].set_ylim([0, 1.1*max(hg[0])])
    axs[int(i>2),i%3].set_title("$\Delta T = "+str(round(dT, 2))+r", \langle \Delta T_{pred}\rangle = $"+str(round(prediction_mean, 1)))


fig.suptitle("Experimental data - performance on testing set")
plt.show()
```
